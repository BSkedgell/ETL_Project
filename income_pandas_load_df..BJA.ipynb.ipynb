```python
income1_file = 'https://usc.data.socrata.com/resource/hs3q-wkhd.json'
#income1_df = pd.read_csv(csv_file)
#income1_df.head()

#income1 = pd.read_csv('https://usc.data.socrata.com/api/views/kygc-fzgm/rows.csv?accessType=DOWNLOAD')

#income1.head()
#creating variable called response, not connecting to database yet 
response = requests.get(income1_file)
```

```python
#for the table location or tract;id, neighborhood, latitude, longitude
neighborhood = []
latitude = []
longitude = []
amount = []
year = []
tract_number = []
variable = []

for record in response.json():
    neighborhood.append(record["neighborhood"])
    latitude.append(record["location"]["coordinates"][0])
    longitude.append(record["location"]["coordinates"][1])
    year.append(record["year"])
    tract_number.append(record["tract_number"])
    variable.append(record["variable"])
    try:
         amount.append(record["amount"])
    except:
         amount.append(0) #this zero (some data in this column didn't have the value) indicates no value
```
